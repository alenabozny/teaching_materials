# Worek słów (Bag of Words, BoW)

Jest to prawdopodobnie najprostszy model reprezentacji tekstu nieustrukturyzowanego w przestrzeni wektorowej. Model przestrzeni wektorowej jest po prostu modelem matematycznym do przedstawiania nieustrukturyzowanego tekstu (lub dowolnych innych danych) jako wektorów numerycznych, tak że każdy wymiar wektora jest określoną cechą/atrybutem. Model BoW przedstawia każdy dokument tekstowy jako wektor liczbowy, gdzie każdy wymiar jest określonym słowem z korpusu, a wartością może być jego częstotliwość w dokumencie, występowanie (oznaczone przez 1 lub 0), a nawet wartości ważone. Nazwa modelu jest taka, ponieważ każdy dokument jest reprezentowany dosłownie jako worek własnych słów, bez względu na kolejność słów, sekwencje i gramatykę.

```python
# KOD POMOCNICZY, DO WYKORZYSTANIA W ĆWICZENIACH

doc_a = 'this document is first document'
doc_b = 'this document is the second document'

bag_of_words_a = doc_a.split(' ')
bag_of_words_b = doc_b.split(' ')

unique_words_set = set(bag_of_words_a).union(set(bag_of_words_b))
print(unique_words_set)
```

# Przykład 2

Skorzystamy z gotowej funkcji wektoryzującej z biblioteki scikit learn. Zauważ, że parametr ngram_range jest ustawiony tak, aby brać pod uwagę pojedyncze słowa (unigramy):

```python
text_transformer = TfidfVectorizer(stop_words='english', ngram_range=(1, 1), lowercase=True, max_features=150000)
```

```python
X_train_text = text_transformer.fit_transform(train['text'])
X_test_text = text_transformer.transform(valid['text'])
```

```python
# Sprawdź, ile próbek i ile cech jest w każdym ze zbiorów:
X_train_text.shape, X_test_text.shape
```

Jako model predykcyjny wykorzystamy **regresję logistyczną**. Nie musisz znać szczegółów implementacyjnych tego algorytmu. Wystarczy na ten moment, że będziesz wiedzieć, iż jest to algorytm klasyfikacyjny, który jest w stanie zwrócic wagi dla poszczególnych cech. Dzięki temu **możemy sprawdzić, jak ważne jest (w naszym przypadku) każde słowo w odniesieniu do przewidywanej etykiety**.

```python
# inicjalizujemy model z parametrami...
logit = LogisticRegression(C=5e1, solver='lbfgs', multi_class='multinomial', random_state=17, n_jobs=4)
```

```python
# ... po czym trenujemy
logit.fit(X_train_text, train['label'])
```

```python
# sprawdźmy, jak dobrze model dopasował się do danych treningowych:
logit.score(X_train_text, train['label'])
```

```python
# Teraz sprawdźmy, jak dobrze model przewiduje etykiety dla danych testowych:

test_preds = logit.predict(X_test_text)

from sklearn.metrics import f1_score
f1_score(np.array(valid['label']), test_preds, 
         labels=['dogs', 'cats', 'fish aquatic pets', 'bunny rabbit central',
       'birds', 'small animals'],
       average='macro')
```

Sprawdź, jak wyliczamy miarę F1 dla zadania klasyfikacji wieloklasowej: https://towardsdatascience.com/micro-macro-weighted-averages-of-f1-score-clearly-explained-b603420b292f


Obiecałam pokazać możliwości objaśniające regresji logistycznej. Dzięki bibliotece eli5 możemy wypisać, dla każdej klasy, jakie cechy są najistotniejsze przy jej przewidywaniu.**Przyjrzyj się wynikom - czy są według Ciebie sensowne?**

```python
eli5.show_weights(estimator=logit, 
                  feature_names= list(text_transformer.vocabulary_),
                 top=(50, 5))
```
